```python
#Bibliotecas
import numpy as np
from scipy import integrate, linalg
import pylab as pl
import matplotlib.pyplot as plt
import scipy.linalg as la
from scipy.integrate import odeint, solve_ivp
from numpy.linalg import matrix_power as mpow
```

```python
#Funci√≥n del oscilador
def oscilador(y,t,omega_c,zeta,m,f0,omega):
  pos, vel = y
  dydt = [vel, f0*np.sin(omega*t)/m - 2.0*zeta*omega_c*vel-omega_c**2*pos]
  return dydt
```

PROBLEMA 1:


```python
#Definimos las constantes
m = 1
f0 = 0
k = 9
omega_c = np.sqrt(k/m)
omega = 0

z1 = 0
z2 = 0.2
z3 = 0.4
z4 = 0.7
z5 = 1
z6 = 2

t = np.linspace(0,3*np.pi,500)
y0 = [-1.0,0.0]

#Obtenemos las soluciones
s1 = odeint(oscilador, y0, t, args=(omega_c, z1, m, f0, omega))
s2 = odeint(oscilador, y0, t, args=(omega_c, z2, m, f0, omega))
s3 = odeint(oscilador, y0, t, args=(omega_c, z3, m, f0, omega))
s4 = odeint(oscilador, y0, t, args=(omega_c, z4, m, f0, omega))
s5 = odeint(oscilador, y0, t, args=(omega_c, z5, m, f0, omega))
s6 = odeint(oscilador, y0, t, args=(omega_c, z6, m, f0, omega))

plt.figure(figsize=(10,5))
plt.title("Posici√≥n contra tiempo para el oscilador amortiguado sin forzamiento")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid()

plt.plot(t, s1[:, 0], label="zeta=0")
plt.plot(t, s2[:, 0], label="zeta=0.2")
plt.plot(t, s3[:, 0], label="zeta=0.4")
plt.plot(t, s4[:, 0], label="zeta=0.7")
plt.plot(t, s5[:, 0], label="zeta=1")
plt.plot(t, s6[:, 0], label="zeta=2")

plt.legend(loc="best")
plt.show()
```
